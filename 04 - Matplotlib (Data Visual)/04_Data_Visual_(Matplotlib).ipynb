***01 - Matplotlib Part 1***

```python
import matplotlib.pyplot as plt
```

```python
plt.show()
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```



> Plot



```python
# FUNCTIONAL
plt.plot(x,y,'r-')    #color
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()    #optional
```



> Subplot



```python
plt.subplot(1,2,1)    # rows, columns
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,b'b')
```



> figure() + add_axes([left, bottom, width, height])



```python
# oo
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  # left, bottom, width, height of the sort of axes. percent
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

***02 - Matplotlib - Part 2***

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)             # axes: array
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
axes
```



> Figure Size and DPI



```python
fig = plt.figure(figsize = (8,2), dpi = 100)      # dpi = dot per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1, figsize = (8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi = 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=(0.1,0.1))
```

***03 - Matplotlib - Part 3***

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = '#FF8C00',lw = 20,alpha = 0.5,ls = '-.',marker='o',markersize = 20,
        markerfacecolor = 'red',markeredgewidth=3,markeredgecolor = 'green')  # RGB Hex Code; alpha:transparency
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
